```python
current_balance = 0.0

def display_balance():
    global current_balance
    print(f"Current balance: {current_balance}")

def deposit():
    global current_balance
    amount = float(input("Enter money to deposit: "))
    if amount > 0:
        current_balance += amount
        print(f"{amount} deposited successfully!")
    else:
        print("Enter a valid amount.")

def withdraw():
    global current_balance
    amount = float(input("Enter money to withdraw: "))
    if 0 < amount <= current_balance:
        current_balance -= amount
        print(f"{amount} withdrawn successfully!")
    else:
        print("Insufficient balance or invalid amount.")

def send_money():
    global current_balance
    amount = float(input("Enter money to send: "))
    if 0 < amount <= current_balance:
        current_balance -= amount
        sum_odd=0
        sum_even=0
        total=0
        card=input("Enter card number:")
        card=card.replace("-","")
        card=card.replace(" ","")
        card=card[::-1]

        for x in card[::2]:
            sum_odd+=int(x)

        for x in card[1::2]:
            x=int(x)*2
            if x>=10:
                sum_even+=((x%10)+1)
            else:
                sum_even+=x

        total=sum_odd+sum_even
        if total%10==0:
            print(f"{amount} sent successfully!")
        else:
            print("Invalid card number, try again!")
    else:
        print("Insufficient balance or invalid amount.")

while True:
    print("\n**********************")
    print("1. Balance remaining")
    print("2. Withdraw money")
    print("3. Deposit money")
    print("4. Send money")
    print("5. Exit")
    print("**********************")
    
    option = input("Select your option: ")

    if option == '1':
        display_balance()
    elif option == '2':
        withdraw()
    elif option == '3':
        deposit()
    elif option == '4':
        send_money()
    elif option == '5':
        print("Exiting...")
        break
    else:
        print("Invalid option. Please try again.")
        
print("Thank you for banking with us!")
```
